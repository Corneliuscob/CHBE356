# pandas (Python Data Analysis Library)

Pandas is a very useful library for doing any kind of data cleaning or data preparation. It also allows us to do some fast data analysis like summary statistics and simple plots.

![iconic trio](../../assets/iconic_trio.png)

Pandas takes in data, like Excel spreadsheets or CSVs, and turns them into a Python object called a **data frame** with rows and objects, which is almost always easier to work with compared to pure Python lists or numpy arrays. You might already be familiar with the concept of data frames from your `R` labs in STAT 251 (Elementary Statistics).

Let's dive right into `pandas` with a simple example:

## 1.  Data Source

We're going to grab weather data from [Environment Canada](http://climate.weather.gc.ca/historical_data/search_historic_data_e.html) for our analysis.

![pandas_logo](../../assets/pandas_logo.png)

To be specific, we'll download July 2018 data for Vancouver, BC as a CSV file from [this link](http://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=51442&Year=2018&Month=07&timeframe=1). If you open the CSV file in Excel, you’ll notice that it contains 16 rows of information and the actual weather data begins at row 17. We'll tell `pandas` this information so it can parse the csv file properly.

P/S: `pandas` will download the .csv data automatically for you using `read_csv`, isn't that cool?

```python
import pandas as pd
source = 'http://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=51442&Year=2017&Month=11&timeframe=1'
```

Let's read our data and store it in a data frame called `df`. We'll tell pandas to skip the first 15 rows.

```python
df = pd.read_csv(source, skiprows=15)
```

Now, we'll take a look at what we have in the first 5 rows using .head().

```python
df.head()
```

To take a look at more rows we can pass a number to `head()`. For example, here's 10 rows:

```python
df.head(10)
```

## 2. Data Cleaning

For the purpose of this tutorial, we'll only use the Date/Time column and the Temp (°C) column for our analysis. Let's copy it to a different data frame.

```python
data = df[['Date/Time', 'Temp (°C)']].copy() # We explicitly tell pandas that we want to make a copy
data.head()
```

We'll need to tell `pandas` what kind of data we have in the data frame. Pandas will try to guess the data type, but if it's unable to, they will be parsed as generic 'objects', as you can see with the dtype:

```python
print(data['Date/Time'].dtype)
print(data['Temp (°C)'].dtype)
```

```python
# We know that the first column is a datetime column:
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
print(data['Date/Time'].dtype)
```

Since the temperature is already parsed correctly as a numeric float, we don't need to change its data type.

## 3. Data Analysis

Let's take a quick look at our data. It's really easy with `pandas`.

```python
%matplotlib inline
data.plot(x='Date/Time', y='Temp (°C)', figsize=(15,6))
```
